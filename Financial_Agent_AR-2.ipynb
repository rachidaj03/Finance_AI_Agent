```python
!pip install -q psycopg2-binary requests

import os, io, requests, textwrap
from typing import List
import psycopg2
from pypdf import PdfReader
from sentence_transformers import SentenceTransformer
from groq import Groq

GROQ_API_KEY  = userdata.get('GROQ_API_KEY')
DB_URL_PSYCO  = "host=localhost port=5432 dbname=ai user=ai password=ai"
TABLE_NAME    = "fin_documents"
EMBED_MODEL   = "sentence-transformers/all-MiniLM-L6-v2"   # 384-dim, runs on CPU
GROQ_MODEL    = "llama-3.3-70b-versatile"
CHUNK_SIZE    = 500    # characters per chunk
CHUNK_OVERLAP = 100

print("Loading embedding model...")
embedder = SentenceTransformer(EMBED_MODEL)
DIMS = embedder.get_sentence_embedding_dimension()
print(f"Embedder ready  |  dims={DIMS}")


def get_conn():
    return psycopg2.connect(DB_URL_PSYCO)

def setup_table():
    """Create pgvector extension + documents table."""
    with get_conn() as conn, conn.cursor() as cur:
        cur.execute("CREATE EXTENSION IF NOT EXISTS vector;")
        cur.execute(f"""
            DROP TABLE IF EXISTS {TABLE_NAME};
            CREATE TABLE {TABLE_NAME} (
                id      SERIAL PRIMARY KEY,
                source  TEXT,
                chunk   TEXT,
                vec     vector({DIMS})
            );
        """)
        conn.commit()
    print(f"Table '{TABLE_NAME}' ready in PostgreSQL")

def download_pdf_text(url: str) -> str:
    print(f"Downloading PDF: {url}")
    r = requests.get(url, timeout=60)
    r.raise_for_status()
    reader = PdfReader(io.BytesIO(r.content))
    pages  = [p.extract_text() or "" for p in reader.pages]
    text   = "\n".join(pages)
    print(f"   → {len(reader.pages)} pages, {len(text):,} chars extracted")
    return text

def chunk_text(text: str, size=CHUNK_SIZE, overlap=CHUNK_OVERLAP) -> List[str]:
    chunks, start = [], 0
    while start < len(text):
        end = start + size
        chunks.append(text[start:end].strip())
        start += size - overlap
    return [c for c in chunks if len(c) > 50]

def store_chunks(chunks: List[str], source: str):
    print(f"   Embedding {len(chunks)} chunks...")
    vecs = embedder.encode(chunks, show_progress_bar=True, batch_size=64)
    with get_conn() as conn, conn.cursor() as cur:
        for chunk, vec in zip(chunks, vecs):
            cur.execute(
                f"INSERT INTO {TABLE_NAME} (source, chunk, vec) VALUES (%s, %s, %s)",
                (source, chunk, vec.tolist())
            )
        conn.commit()
    print(f"Stored {len(chunks)} chunks from: {source}")


def retrieve(query: str, top_k: int = 5) -> List[str]:
    q_vec = embedder.encode(query).tolist()
    with get_conn() as conn, conn.cursor() as cur:
        cur.execute(f"""
            SELECT chunk
            FROM   {TABLE_NAME}
            ORDER  BY vec <-> %s::vector
            LIMIT  %s
        """, (q_vec, top_k))
        rows = cur.fetchall()
    return [r[0] for r in rows]

groq_client = Groq(api_key=GROQ_API_KEY)

def ask(question: str, top_k: int = 5):
    print(f"\n❓ {question}")
    contexts = retrieve(question, top_k=top_k)
    context_block = "\n\n---\n\n".join(contexts)

    system_prompt = textwrap.dedent("""
        You are a precise financial analyst assistant.
        Answer ONLY using the provided context.
        If the context does not contain enough information, say so clearly.
        Be concise and structured.
    """).strip()

    user_prompt = f"""Context:\n{context_block}\n\nQuestion: {question}"""

    response = groq_client.chat.completions.create(
        model=GROQ_MODEL,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user",   "content": user_prompt},
        ],
        temperature=0.2,
        max_tokens=1024,
    )
    answer = response.choices[0].message.content
    print(f"\n💡 {answer}\n")
    return answer
```

# Evaluation : LLM as a judge

```python
from textwrap import dedent
from agno.agent import Agent
from agno.models.groq import Groq

class RAGEvaluator:
    def __init__(self):
        self.evaluator = self._initialize_evaluator()

    def _initialize_evaluator(self):
        return Agent(
            model=Groq(id="llama-3.1-8b-instant"),
            description=dedent("""\
                You are an expert RAG system evaluator with deep expertise in:
                - Information retrieval quality assessment
                - Response accuracy evaluation
                - Source attribution verification
                - Context relevance analysis
                - Natural language generation evaluation
            """),
            instructions=dedent("""\
                Evaluate the RAG system output based on these key metrics:

                1. Faithfulness (1-5):
                   - How accurately does the response reflect the source documents?
                   - Are there any hallucinations or incorrect statements?
                   - Does it maintain factual consistency?

                2. Context Relevance (1-5):
                   - Are the retrieved passages relevant to the query?
                   - Is important context missing?
                   - Is irrelevant information included?

                3. Answer Completeness (1-5):
                   - Does the response fully address the query?
                   - Are all key aspects covered?
                   - Is the level of detail appropriate?

                4. Source Attribution (1-5):
                   - Are sources properly cited?
                   - Is it clear which information comes from where?
                   - Can claims be traced back to sources?

                5. Response Coherence (1-5):
                   - Is the response well-structured?
                   - Does it flow logically?
                   - Is it easy to understand?

                Provide specific examples and explanations for each score.
            """),
            expected_output=dedent("""\
                # RAG Evaluation Report

                ## Overview
                Query: {query}
                Response Length: {n_chars} characters

                ## Metric Scores

                ### Faithfulness: {score}/5
                - Justification:
                - Examples:
                - Areas for Improvement:

                ### Context Relevance: {score}/5
                - Justification:
                - Examples:
                - Areas for Improvement:

                ### Answer Completeness: {score}/5
                - Justification:
                - Examples:
                - Areas for Improvement:

                ### Source Attribution: {score}/5
                - Justification:
                - Examples:
                - Areas for Improvement:

                ### Response Coherence: {score}/5
                - Justification:
                - Examples:
                - Areas for Improvement:

                ## Overall Score: {total}/25

                ## Key Recommendations
                1. {rec1}
                2. {rec2}
                3. {rec3}

                ## Summary
                {final_assessment}
            """),
            markdown=True,
        )

    def evaluate(self, query: str, response: str, context: list, stream: bool = True):
        """
        Evaluate a RAG system's response

        Args:
            query (str): Original user query
            response (str): RAG system's response
            context (list): Retrieved passages used for the response
            stream (bool): Whether to stream the evaluation output
        """
        evaluation_prompt = f"""
        Please evaluate this RAG system output:

        QUERY:
        {query}

        RETRIEVED CONTEXT:
        {' '.join(context)}

        RESPONSE:
        {response}

        Provide a detailed evaluation following the metrics and format specified.
        """

        return self.evaluator.print_response(evaluation_prompt, stream=stream)

evaluator = RAGEvaluator()
print("LLM-as-a Judge Evaluator initialized successfully!")
```

```python
# Example evaluation.

query = "What are the key features of transformer models?"
context = [
    "Transformer models use self-attention mechanisms to process input sequences.",
    "Key features include parallel processing and handling of long-range dependencies."
]
response = "Transformer models are characterized by their self-attention mechanism..."
evaluator.evaluate(query, response, context)
```

# Final Pipeline Generation

```python
def run_agent(agent, prompt: str) -> str:
    result = agent.run(prompt)
    if hasattr(result, "content"):
        return result.content
    return str(result)
```

```python
def run_financial_pipeline(
    user_query: str,
    pdf_url: str = None,
    top_k: int = 5
):
    output = {
        "user_query": user_query,
        "rag_context": None,
        "research_context": None,
        "stock_context": None,
        "final_report": None
    }
    if pdf_url:
        print("Setting up table and indexing PDF...")
        setup_table()
        raw_text = download_pdf_text(pdf_url)
        chunks = chunk_text(raw_text)
        store_chunks(chunks, source=pdf_url)
        print(f"Indexed {len(chunks)} chunks from PDF")

    print("Retrieving relevant chunks...")
    retrieved = retrieve(user_query, top_k=top_k)
    rag_context = ask(user_query)
    output["rag_context"] = rag_context

    print("Running research agent...")
    research_context = run_agent(research_agent, user_query)
    output["research_context"] = research_context

    stock_context = run_agent(stock_agent,user_query)
    output["stock_context"] = stock_context

    print("Generating final report...")
    report_prompt = f"""
    USER REQUEST:
    {user_query}

    INTERNAL DOCUMENT CONTEXT:
    {rag_context}

    EXTERNAL RESEARCH CONTEXT:
    {research_context}

    STOCK / MARKET ENRICHMENT:
    {stock_context}

    Build the final professional report.
    """
    final_report = run_agent(report_generator_agent, report_prompt)
    output["final_report"] = final_report

    return output
```

```python
result = run_financial_pipeline(
    user_query="OCP want to invest in some Cobalt mines in Canada, and want some financial advice"
)
```
